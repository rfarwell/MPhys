This code splits our dataset into the training, testing and validation sets. 70% training, 15% testing, 15% validation. 
This works for our data set of 400 patients in which, at a timeframe of 1.5years, has 200 alive and 200 dead. 
The iterates 200 times (length of alive/dead arrays) each time assigning a randomly generated integer to "random_int" (with a specified range) and then using this integer as an index to remove the relevant element from both alive and dead arrays and add to the newly assigned arrays (validation, training, testing).

```python
import os
import numpy as np
import random
```

```python
clinical_data_filepath = '/Volumes/Extreme_SSD/MPhys/TCIA_Data/NSCLC-Radiomics/NSCLC-Radiomics-Clinical-Data.csv' #defining the filepath to the clinical data
clinical_data = np.genfromtxt(clinical_data_filepath,comments = '%', delimiter = ',', dtype = str) #opening the file and generating an array from it

patient_names = clinical_data[:,0]
timepoints = clinical_data[:,8]
timepoints = timepoints.astype(np.int)
dead_statuses = clinical_data[:,9]
dead_statuses = dead_statuses.astype(np.int0)

# print(patient_names)
# print(len(patient_names))
# print(timepoints)
# print(len(timepoints))
# print(dead_statuses)
# print(len(dead_statuses))

```

```python
"""Making a training array (70%) of full data set"""
alive_array = []
dead_array = []

for i in range(len(patient_names)):
  temp_patient_name = patient_names[i]
  temp_timepoint = timepoints[i]
  temp_dead_statuses = dead_statuses[i]
  if temp_dead_statuses == 1 :
    if temp_timepoint > 365*1.5 :
      temp = [temp_patient_name, 0]
      alive_array.append(temp)
    elif temp_timepoint < 365*1.5 :
      temp = [temp_patient_name, 1]
      dead_array.append(temp)
  
  if temp_dead_statuses == 0 :
    if temp_timepoint > 365*1.5 :
      temp = [temp_patient_name, 0]
      alive_array.append(temp)
    elif temp_timepoint < 365*1.5 :
      print("Alive before timeframe")

print(f"alive_array: {alive_array}")
print(f"The length of alive_array is {len(alive_array)}")
print(f"dead_array: {dead_array}")
print(f"The length of dead_array is {len(dead_array)}")

original_alive_array_length = len(alive_array)
range_of_rand = 199
training_array = []
training_array_length = original_alive_array_length * 0.7

for i in range(int(training_array_length)) :
  print(f"Interation number: {i}")
  print(f"Range of random integer is  {range_of_rand}")
  random_int = random.randint(0,range_of_rand)
  print(f"Random integer being used is {random_int}")
  range_of_rand = range_of_rand - 1
  #print(f"range_of_rand = {range_of_rand}")
  print(random_int)
  print(f"Appending {alive_array[random_int]} to training_array")
  training_array.append(alive_array[random_int])
  print(f"Appending {dead_array[random_int]} to training_array")
  training_array.append(dead_array[random_int])
  del alive_array[random_int]
  del dead_array[random_int]
  print(f"Length of alive_array is {len(alive_array)}")
  print(f"Length of dead_array is {len(dead_array)}")
  print("=====================================================================")

print(f"Training array : {training_array}")
print(f"Length of training array is {len(training_array)}")

  

training_dead_counter = 0
training_alive_counter = 0

for i in range(len(training_array)) :
  if training_array[i][2] == 1 :
    training_dead_counter += 1
  elif training_array[i][2] == 0 :
    training_alive_counter += 1
  
print(f"Training_dead_counter = {training_dead_counter}")
print(f"Training_alive_counter = {training_alive_counter}")
```

```python
testing_array = []
testing_array_length = original_alive_array_length * 0.15

for i in range(int(testing_array_length)) :
  print(f"Interation number: {i}")
  print(f"Range of random integer is  {range_of_rand}")
  random_int = random.randint(0,range_of_rand)
  print(f"Random integer being used is {random_int}")
  range_of_rand = range_of_rand - 1
  #print(f"range_of_rand = {range_of_rand}")
  print(random_int)
  print(f"Appending {alive_array[random_int]} to testing_array")
  testing_array.append(alive_array[random_int])
  print(f"Appending {dead_array[random_int]} to testing_array")
  testing_array.append(dead_array[random_int])
  del alive_array[random_int]
  del dead_array[random_int]
  print(f"Length of alive_array is {len(alive_array)}")
  print(f"Length of dead_array is {len(dead_array)}")
  print("=====================================================================")

print(f"Testing array : {testing_array}")
print(f"Length of testing array is {len(testing_array)}")


testing_dead_counter = 0
testing_alive_counter = 0

for i in range(len(testing_array)) :
  if testing_array[i][2] == 1 :
    testing_dead_counter += 1
  elif testing_array[i][2] == 0 :
    testing_alive_counter += 1
  
print(f"Testing_dead_counter = {testing_dead_counter}")
print(f"Testing_alive_counter = {testing_alive_counter}")
```

```python
validation_array = []
validation_array_length = original_alive_array_length * 0.15

for i in range(int(validation_array_length)) :
  print(f"Interation number: {i}")
  print(f"Range of random integer is  {range_of_rand}")
  random_int = random.randint(0,range_of_rand)
  print(f"Random integer being used is {random_int}")
  range_of_rand = range_of_rand - 1
  #print(f"range_of_rand = {range_of_rand}")
  print(random_int)
  print(f"Appending {alive_array[random_int]} to validation_array")
  validation_array.append(alive_array[random_int])
  print(f"Appending {dead_array[random_int]} to validation_array")
  validation_array.append(dead_array[random_int])
  del alive_array[random_int]
  del dead_array[random_int]
  print(f"Length of alive_array is {len(alive_array)}")
  print(f"Length of dead_array is {len(dead_array)}")
  print("=====================================================================")

print(f"Validation array : {validation_array}")
print(f"Length of validation array is {len(validation_array)}")


validation_dead_counter = 0
validation_alive_counter = 0

for i in range(len(validation_array)) :
  if validation_array[i][2] == 1 :
    validation_dead_counter += 1
  elif validation_array[i][2] == 0 :
    validation_alive_counter += 1
  
print(f"Validation_dead_counter = {validation_dead_counter}")
print(f"Validation_alive_counter = {validation_alive_counter}")
```
